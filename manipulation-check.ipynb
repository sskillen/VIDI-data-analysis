```python
import pandas as pd
# Assuming your DataFrame is named 'df'

cols_misinfo = [col for col in df_values.columns if col.startswith('misinfo')]

df_values[cols_misinfo] = df_values[cols_misinfo].apply(pd.to_numeric, errors='coerce')
```

```python
df_values[cols_misinfo].mean()
```

```python
conditions = ['inciv_', 'misin_', 'us_them_', 'control_', 'intol_']
cols_dict = {}

for condition in conditions:
    cols = [col for col in df_values.columns if col.startswith(condition)]
    cols_dict[condition] = cols
    df_values[cols] = df_values[cols].apply(pd.to_numeric, errors='coerce')

# Access them like:
print(df_values[cols_dict['us_them_']].mean())
#print(df_values[cols_dict['inciv_']].mean())
```

```python
grouped = {}
for prefix, cols in cols_dict.items():
    groups = {}
    for num in range(1, 8):
        if prefix == 'us_them_':
            # For 'us_them_', the number is at index [2] due to underscore in the prefix
            matching_cols = [col for col in cols if len(col.split('_')) > 2 and col.split('_')[2] == str(num)]
        else:
            # For other prefixes, the number is at index [1]
            matching_cols = [col for col in cols if len(col.split('_')) > 1 and col.split('_')[1] == str(num)]
        groups[num] = matching_cols
    grouped[prefix] = groups
```

```python
for prefix, groups in grouped.items():
    print(f"\n{prefix}:")
    for num, cols in sorted(groups.items()):
        if cols:
            means = df_values[cols].mean()
            max_mean = means.max()
            top_cols = means[means >= max_mean - 0.10].sort_values(ascending=False)
            print(f"  Group {num}: {top_cols.to_dict()}")
```
